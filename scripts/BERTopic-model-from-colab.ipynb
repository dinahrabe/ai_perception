# Topic models

## GPU accelerated

```python
from bertopic import BERTopic
from cuml.cluster import HDBSCAN
from cuml.manifold import UMAP
```

```python
###GPU

# Create instances of GPU-accelerated UMAP and HDBSCAN
umap_model = UMAP(n_components=5, n_neighbors=15, min_dist=0.0, random_state=1505)
hdbscan_model = HDBSCAN(min_samples=10, gen_min_span_tree=True)

# Pass the above models to be used in BERTopic
topic_model = BERTopic(verbose=True, umap_model=umap_model, hdbscan_model=hdbscan_model, nr_topics=50)
topics, probs = topic_model.fit_transform(text)
```

```python
topic_model.save("/content/drive/MyDrive/text-as-data/model_50t")
```

## Online learning

```python
del topic_model
del topics
```

```python
#doc chunks of 1000
doc_chunks = [text[i:i+1000] for i in range(0, len(text), 1000)]
```

```python
len(doc_chunks)
```

```python
from sklearn.cluster import MiniBatchKMeans
from sklearn.decomposition import IncrementalPCA
from bertopic.vectorizers import OnlineCountVectorizer

# Prepare sub-models that support online learning
umap_model = IncrementalPCA(n_components=5)
cluster_model = MiniBatchKMeans(n_clusters=50, random_state=0)
vectorizer_model = OnlineCountVectorizer(stop_words="english", decay=.01)
```

```python
from bertopic import BERTopic

topic_model = BERTopic(umap_model=umap_model, hdbscan_model=cluster_model, vectorizer_model=vectorizer_model)

# Incrementally fit the topic model by training on 1000 documents at a time and track the topics in each iteration
topics = []

for docs in doc_chunks:
    topic_model.partial_fit(docs)
    topics.extend(topic_model.topics_)

topic_model.topics_ = topics
```

```python
topic_model.save("/content/drive/MyDrive/text-as-data/model_online")
```

```python
topics = [3, 13, 18, 22, 24, 25, 27, 29, 32, 33, 40, 42]

topic_plot = topic_info.loc[topic_info['Topic'].isin(topics)]
```

```python
docs = topic_model.get_representative_docs(topic=3)[0]
```

```python
representative_topics = {}

for topic in topics:
  representative_topics[topic] = [topic_model.get_representative_docs(topic=topic)[0]]
```

```python
rep_top = pd.DataFrame(representative_topics).melt(var_name="Topic", value_name='representative_sentence')
```

```python
topic_plot = topic_plot.merge(rep_top)
topic_plot = topic_plot.sort_values(by=["Count"], ascending=True)
topic_plot["constant"] = 0.2
topic_plot["constant_2"] = 0.9
```

```python
topic_plot.head()
```

```python
#plot topic size, topic name and indicative sentence

import plotly.express as px
import kaleido

fig = px.scatter(topic_plot, x="constant", y="Topic", text="Name", size="Count", template="plotly_white", size_max=40)
#fig.add_scatter(x=topic_plot["constant_2"], y=topic_plot["Topic"], text=topic_plot["representative_sentence"])

fig.update_traces(textposition='middle right', marker=dict(symbol="square"))

fig.update_xaxes(range=[0.1, 2.8])
fig['layout']['yaxis']['autorange'] = "reversed"

txt=[]
counter=0
for i in topic_plot['representative_sentence'].tolist():
    t=dict(x=topic_plot['constant_2'].values[counter],y=topic_plot['Topic'].values[counter],text=topic_plot['representative_sentence'].values[counter], showarrow=False, xanchor="left")
    txt.append(t)
    counter+=1

fig.update_layout(
    height=800,
    title_text='AI-related topics and their representative sentences',
    xaxis=dict(showgrid=False, visible= False),
    yaxis=dict(showgrid=False, visible= False),
    annotations=txt
)

fig.show()
```
